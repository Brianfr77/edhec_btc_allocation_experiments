## Inspect Required Outputs

```python
import json
import pandas as pd
from IPython.display import display

RUN_DIR = PROJECT_ROOT / 'outputs' / 'part1_btc_macro_state' / RUN_ID
RESULTS_DIR = RUN_DIR / 'results'

required_outputs = [
    'validation_summary.json',
    'btc_asset_diagnostics.csv',
    'btc_source_validation.csv',
    'return_correlation_matrix.csv',
    'rolling_correlations_52w.csv',
    'pca_explained_variance.csv',
    'pca_loadings.csv',
    'hmm4_state_labels.csv',
    'hmm4_state_profiles.csv',
    'hmm4_transition_matrix.csv',
    'model_robustness_summary.csv',
    'state_conditioned_btc_summary.csv',
    'state_conditioned_correlations.csv',
    'data_lineage.csv',
    'variable_dictionary.csv',
    'model_assumption_audit.json',
    'state_labeling_rationale.csv',
    'hmm_initialization_audit.csv',
    'methodology_audit.md',
]

for name in required_outputs:
    path = RESULTS_DIR / name
    print(f'{name}: {path.exists()}')

with open(RUN_DIR / 'run_manifest.json', 'r', encoding='utf-8') as handle:
    manifest = json.load(handle)
print(json.dumps(manifest['sample'], indent=2))
```
