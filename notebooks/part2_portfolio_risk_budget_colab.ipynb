## Inspect Required Outputs

```python
import json
import pandas as pd
from IPython.display import display

RUN_DIR = PROJECT_ROOT / 'outputs' / 'part2_portfolio_risk_budget' / RUN_ID
RESULTS_DIR = RUN_DIR / 'results'

required_outputs = [
    'input_validation_summary.json',
    'baseline_portfolio_weights.csv',
    'fixed_btc_portfolio_weights.csv',
    'erc_optimization_diagnostics.csv',
    'portfolio_return_series.csv',
    'portfolio_performance_summary.csv',
    'full_sample_risk_contributions_vol.csv',
    'full_sample_risk_contributions_cvar.csv',
    'btc_risk_budget_summary.csv',
    'state_conditioned_portfolio_summary.csv',
    'state_conditioned_btc_risk_contributions.csv',
    'state_conditioned_drawdowns.csv',
    'methodology_audit.md',
    'data_lineage.csv',
    'portfolio_weight_dictionary.csv',
    'risk_contribution_methodology.json',
    'model_assumption_audit.json',
    'output_validation_summary.json',
]

for name in required_outputs:
    path = RESULTS_DIR / name
    print(f'{name}: {path.exists()}')

with open(RUN_DIR / 'run_manifest.json', 'r', encoding='utf-8') as handle:
    manifest = json.load(handle)
print(json.dumps(manifest['sample'], indent=2))
print(json.dumps(manifest['output_validation'], indent=2))
```
